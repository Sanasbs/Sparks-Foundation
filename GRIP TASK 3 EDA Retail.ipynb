#  STEP 4: Exploratory Data Analysis (EDA)

To focus on the weak area, creating a separate loss dataframe.


```python
 loss=ssdata[ssdata['Profit'] < 0]                      ##creating loss dataframe
```

```python
loss
```

```python
loss.shape
```

```python
loss.describe()                                           ##statistical details of loss dataframe
```

```python
Total_loss=np.negative(loss['Profit'].sum())
print("Total loss = %.2f" %Total_loss)                      ## total loss calculation 
```

```python
loss.groupby(by='Segment').sum()                           ##sum of segment-wise loss 
```

```python
loss.groupby(by='Sub-Category').sum()                      ##sum of sub-category wise loss 
```

```python
loss['Sub-Category'].value_counts()                     ## loss count for sub-categories
```

```python
loss.groupby(by='City').sum().sort_values('Profit',ascending=True).head(15)     ##cities with highest loss
```

```python
loss.sort_values(['Sales'],ascending=True).groupby(by='Category').mean()      ## average of different categories
```

```python
ssdata.sort_values(['Segment'],ascending=True).groupby('Segment').sum()     ## sum of different sengments
```

```python
ssdata.groupby(by='Region').sum()                                           ## sum of different regions
```

# STEP 5: Visualization

```python
plt.figure(figsize=(25,5))
plt.bar(loss['Sub-Category'],loss['Sales']);
plt.xlabel('Sub_Category');                                              ## visualization of sub-category vs sales (loss)
plt.ylabel('Sales');
```

```python
plt.figure(figsize=(25,5))
plt.bar(ssdata['Sub-Category'],ssdata['Sales']);
plt.xlabel('Sub_Category');                                         ## visualization of sub-category vs sales (original dataset)
plt.ylabel('Sales');
```

```python
plt.figure(figsize=(25,5))
plt.bar(ssdata['Sub-Category'],ssdata['Discount']);
plt.xlabel('Sub_Category');                                      ## visualization plot of sub-category vs Discount
plt.ylabel('Discount');
```

```python
plt.figure(figsize=(20,5))
plt.bar(ssdata['Ship Mode'],ssdata['Sales']);                   ## visualization plot of ship mode vs sales
plt.xlabel('Ship Mode');
plt.ylabel('Sales');
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=ssdata.Segment)                                 ## visualization plot of sengments and its count
plt.show();
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=ssdata['Sub-Category'])                          ## visualization plot of sub-category and its count
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=ssdata['Region'])                               ## seaborn plot of region and its count
plt.xticks(rotation=90)
plt.show()
```
